```python
# K means clustering is an unsupervised learning algorithm that will attempt to similar clusters together from your data
#It is mainly used in clustering similar documents
#clustoring customers based on similar features

from sklearn.datasets import make_blobs
#The make_blobs functions is used to generate synthetic datasets for clustering and classification tasks
#This function will create clusters of data points with gaussian distribution

```

```python
#creating random dataset
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=101)
data
```

```python
#n_samples=Total number of points equally divided among clusters
#n_features=It indicated the number of features(columns)
#centers=It determine number of clusters to be generated
#cluster_std=It sets the standard deviation of the clusters .High value makes the clusters to spread out 
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:, 0],x[:, 1],c=y,cmap='rainbow',s=250,edgecolor='black')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("scatter plot for k means")
```

```python
data[0].shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
kmeans
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(8,5))
ax1.set_title('K means')#predicted data
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")#original data
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
